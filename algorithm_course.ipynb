```python
def task_4(K1,M,K2,P2,N2):
    P1,N1 = 0,0
    if ((M * (P2 -1) + N2 - 1) == 0):
        return '{} {}'.format(0,1)
     
    n = K2 // (M * (P2 -1) + N2 - 1)
    if (K2 > (P2 - 1)*n*M + n*(N2 - 1) & K2 < (P2 - 1)*n*M + n*N2 + 1) is False:
        return '{} {}'.format(-1,-1)
    P1 = (K1 // (M * n)) + 1
    N1 = ((K1 - (P1 - 1) * M * n) // n) + 1
    return '{} {}'.format(P1,N1)
```

```python
def solve_system(a, b, c, d, e, f):
    determinant = a * d - b * c
    determinant_x = e * d - b * f
    determinant_y = a * f - e * c
    if determinant != 0:
        x = determinant_x / determinant
        y = determinant_y / determinant
        print('{} {} {}'.format(2, round(x, 5), round(y, 5)))
        return 
    else:
        if determinant_x == 0 and determinant_y == 0:
            if (a == 0 and c == 0) and (b == 0 and d == 0):
                if e != 0 or f != 0:
                    print('0')
                else:
                    print('5')
            elif a == 0 and c == 0:
                if b != 0:
                    y = e / b
                else:
                    y = f / d
                print('{} {}'.format(4, round(y, 5)))
            elif b == 0 and d == 0:
                if a != 0:
                    x = e / a
                else:
                    x = f / c
                print('{} {}'.format(3, round(x, 5)))
            else:
                if b != 0:
                    bi = e / b
                    k = -a / b
                else:
                    bi = f / d
                    k = -c / d
                print('{} {} {}'.format(1, round(k, 5), round(bi, 5)))
        else:
            print('0')

    
# Test case
solve_system(1, 0, 0, 1, 3, 3)
solve_system(1, 1, 2, 2, 1, 2)
solve_system(0, 2, 0, 4, 1, 2)
solve_system(0, 4, 4, 0, 1, 2)
```

```python
def task_7(n,k,m):
    count = 0
    mod = 0
    if k > n:
        return 0
    if m > k: 
        return 0
    while n >= k:
        count += (n // k) * (k // m)
        mod = (n // k) * (k % m)
        n = (n % k) + mod
    return count
    
```

```python
n,k,m = list(map(int, input().split()))
print(task_7(n,k,m))
```

<h1>Task 8 Queue<h1/>

```python
class Queue:
    def __init__(self):
        self.queue = list() 
 
    def add_element(self,val): 
        if val not in self.queue: 
            self.queue.insert(0,val) 
            return True 
        return False 
    
    def pull_element(self):
        if len(self.queue)>0:
            return self.queue.pop() 
        return ("Queue is Empty")
    
    def is_empty(self):
        if self.size():
            print("Queue is not Empty")  
        else:
            print("Queue is Empty") 
    
    def queue_print(self):
        for element in self.queue:
            print(element,end=' ')
        print()
        
    def clean_queue(self):
        self.queue.clear()
    
    def size(self): 
        return len(self.queue) 
```

```python
TheQueue = Queue() 
TheQueue.is_empty()
TheQueue.add_element("Apple") 
TheQueue.add_element("Mango") 
TheQueue.add_element("Guava") 
TheQueue.add_element("Papaya") 
print("The length of Queue: ",TheQueue.size()) 
TheQueue.queue_print()
TheQueue.pull_element()
TheQueue.queue_print()
TheQueue.is_empty()
print("The length of Queue: ",TheQueue.size()) 
```
